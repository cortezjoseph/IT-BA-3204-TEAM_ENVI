```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import plotly as py
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import plotly.graph_objs as go

#Changing of fonts
plt.rcParams["font.family"] = "serif"
```

```python
url = 'https://raw.githubusercontent.com/cortezjoseph/IT3204_TreeRestrial/main/Datasets/monthly%20climatology.csv'
df = pd.read_csv(url)
newdf = df.drop([0,1,2], axis=0)
newdf.rename(columns={'Monthly Climatology': 'Category','Unnamed: 1': 'MinTemperature', 'Unnamed: 2': 'MeanTemperature', 'Unnamed: 3': 'MaxTemperature', 'Unnamed: 4': 'Precipitation'}, inplace=True)
display(newdf.iloc[0:])
     
```

```python
#converting datatype
newdf['MinTemperature']= newdf.MinTemperature.astype(float)
newdf['MeanTemperature']= newdf.MeanTemperature.astype(float)
newdf['MaxTemperature']= newdf.MaxTemperature.astype(float)
newdf['Precipitation']= newdf.Precipitation.astype(float)

#getting all values from the dataframe/csv
Category = newdf["Category"]
MinTemperature = newdf["MinTemperature"]
MeanTemperature = newdf["MeanTemperature"]
MaxTemperature = newdf["MaxTemperature"]
Precipitation = newdf["Precipitation"]

#size of the bar graph
f = plt.figure()
f.set_figwidth(23)
f.set_figheight(10)
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#Setting labels
plt.ylabel("Monthly Climatology" , fontsize=15)
plt.xlabel("Months", fontsize=15)
plt.title("\n Summary of Monthly Climatology in the Philippines ",  fontweight="bold" , fontsize=20)

#creating visualizations 
xpos = np.arange(len(Category))
widthSize=0.25
plt.bar(xpos, MinTemperature, color="#FAAA17", width = widthSize, label='Minimun Temperature')
plt.bar(xpos+0.3, MeanTemperature, color="#AFD43F", width = widthSize, label='Mean Temperature')
plt.bar(xpos+0.6, MaxTemperature, color="#EBC590", width = widthSize, label='Maximum Temperature')

plt.xticks(xpos+0.3, ('JAN', 'FEB', 'MAR', 'APR','MAY','JUN', 'JUL', 'AUG', 'SEP','OCT','NOV','DEC'), fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
#Saving visualizations in TIFF file with 300 dpi
plt.savefig('SummaryWage.tiff', transparent=True, dpi=300)
plt.show()
     
```
